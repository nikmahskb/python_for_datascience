```python
#@title Check duplicate & handle duplicate

## Identification
toyota_data.duplicated()
```

```python
#@title Copy Data

toyota_data2 = toyota_data.copy()
toyota_data2
```

```python
## Add new column duplicated

toyota_data2['duplicated'] = toyota_data2.duplicated()
toyota_data2
```

```python
## Drop Duplicate

TD_new = toyota_data.drop_duplicates()
TD_new
```

```python
#@title Filter Dataframe
```

```python
#@title Filter with logic operator

data_with_mileage_more_than_50k_mile = toyota_data[toyota_data.mileage >= 50000]
data_with_mileage_more_than_50k_mile
```

```python
#@title Multiple Filter

multiple_filter = toyota_data[(toyota_data.mileage >= 50000) & (toyota_data.transmission == 'Automatic') & (toyota_data.year < 2010) & (toyota_data.engineSize == 2.0)]
multiple_filter
```

```python
#@title Filter with isin

fuelType = ['Petrol', 'Diesel']

spesific_model = toyota_data[toyota_data.fuelType.isin(fuelType)]
# spesific_model = toyota_data[toyota_data.year.isin([2010, 2011])]
spesific_model
```

```python
#@title Filter with str accessor

# toyota_data[toyota_data.fuelType.str.startswith('Di')]
toyota_data[toyota_data.fuelType.str.endswith('ol')]
```

```python
#@title Filter with nlargest & nsmallest

#Mengambil 10 mobil dengan  harga termahal
toyota_data.nlargest(10, 'price')
```

```python
#Mengambil 5 mobil dengan mileage terpendek

toyota_data.nsmallest(3, 'mileage')
```

```python
#@title Filter with Query

toyota_data.query('transmission == "Semi-Auto" and year <= 2020 and tax >= 120')
```

```python
#@title Group Dataframe

toyota_data.groupby('transmission').agg({"mileage": "sum"})
```

```python
toyota_data.groupby(['transmission', 'year']).sum()[['tax']]
```

```python
#@title Replace Dataframe

# Replace Values in Column
# replace(oldvalue, newvalue)

toyota_data['transmission'] = toyota_data['transmission'].replace('Manual','Berkopling')
print(toyota_data)
```

```python
# Replace Multiple Values in Column

toyota_data['fuelType'] = toyota_data['fuelType'].replace(['Petrol', 'Diesel'], 'Pertamax')
print(toyota_data)
```
