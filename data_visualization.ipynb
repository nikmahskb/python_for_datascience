```python
#@title Recursion

# Fungsi rekursif yang dioptimalkan untuk menghitung jumlah laki-laki dan perempuan
def count_gender_optimized(data, index=None, male_count=0, female_count=0):
    # Tentukan index jika None (start dari akhir data)
    if index is None:
        index = len(data) - 1

    # Basis case: Jika index sudah di bawah 0, kembalikan jumlah laki-laki dan perempuan
    if index < 0:
        return f'Jumlah siswa laki-laki: {male_count}, Jumlah siswa perempuan: {female_count}'

    # Ambil jenis kelamin siswa pada indeks saat ini
    gender = data[index]

    # Klasifikasi jenis kelamin
    if gender == 'M':
        male_count += 1
    elif gender == 'F':
        female_count += 1

    # Panggil fungsi rekursif pada indeks sebelumnya
    return count_gender_optimized(data, index - 1, male_count, female_count)
```

```python
jk_stud = student['sex']
count_gender_optimized(jk_stud)
```

```python
#@title First Class

grade1 = student['G1']
grade2 = student['G2']
grade3 = student['G3']
join5 = np.dstack((grade1, grade2, grade3))

def calculate_average(scores):
    scores = data
    return sum(scores) / len(scores)

def process_students(data, operation):
    return [operation(student) for student in data]   # melakukan looping calculate_average

def analyze_exam_scores(data):
    # Menghitung nilai rata-rata untuk setiap siswa
    averages = process_students(data, calculate_average)

    # Mengembalikan hasil
    return averages

results = analyze_exam_scores(join5)


print("Rata-rata nilai setiap siswa: \n", results)
```

# **Numpy**

```python
#@title Statistik

grade1 = student['G1']

mean = np.mean(grade1)
print("Rata-rata (mean):", mean)

median = np.median(grade1)
print("Median:", median)

std_deviation = np.std(grade1)
print("Deviasi Standar (standard deviation):", std_deviation)

variance = np.var(grade1)
print("Variansi (variance):", variance)

minimum = np.min(grade1)
maximum = np.max(grade1)
print("Nilai Minimum:", minimum)
print("Nilai Maksimum:", maximum)

# Menghitung kuartil pertama (Q1) dan kuartil ketiga (Q3)
q1 = np.percentile(grade1, 25)
q3 = np.percentile(grade1, 75)
print("Kuartil Pertama (Q1):", q1)
print("Kuartil Ketiga (Q3):", q3)


# Menghitung rentang interkuartil (IQR)
iqr = q3 - q1
print("Rentang Interkuartil (IQR):", iqr)
```

```python
#@title Join

grade1 = student['G1']
grade2 = student['G2']
grade3 = student['G3']

join1 = np.concatenate((grade1, grade2, grade3))
join2 = np.stack((grade1, grade2, grade3))
join3 = np.hstack((grade1, grade2, grade3))
join4 = np.vstack((grade1, grade2, grade3))
join5 = np.dstack((grade1, grade2, grade3))

print(f"numpy concatenate: \n {join1} \n")

print(f"numpy stack: \n {join2} \n")

print(f"numpy hstack: \n {join3} \n")

print(f"numpy vstack: \n {join4} \n")

print(f"numpy dstack: \n {join5} \n")
```
